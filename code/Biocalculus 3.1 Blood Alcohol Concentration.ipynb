**Measuring the rate of increase in Blood Alcohol Concentration** by 156 in BioCalculus by Stewart and Day. Reproducing Figure 1:

```python
var('C')
def C(t):
    return 0.0225*t*exp(-0.0467*t)
```

```python
var('t')
plot(C,(t,0,90))
```

```python
var('AvChangeC10')
def AvChangeC10(t2):
    return (C(t2)-C(10))/(t2-10)
```

```python
AvChangeC10(11)
```

```python
tvec = [11, 10.5, 10.1, 10.01]
data=[(t,AvChangeC10(t)) for t in tvec]
data
```

```python
tvec = [9, 9.5, 9.9, 9.99]
data2=[(t,AvChangeC10(t)) for t in tvec]
data2
```
