**Generating the logistic sequence in SageMath**

Example 8. p97 in BioCalculus by Stewart and Day. 

c = 2.8 and $x_0 = 0.8$ 

```python
var('a')
def a(n):
    if n==0:
            return 0.8
    else:
            return 2.8*a(n-1)*(1-a(n-1))
```

```python
data=[(n,a(n)) for n in range(17)]
data
```

**FIGURE 4**

```python
list_plot(data, color = "red")
```
