```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function with a jump discontinuity
def f(x):
    if x < 0:
        return 0
    else:
        return 1

# Generate x values
x_values = np.linspace(-2, 2, 400)

# Generate y values using the function
y_values = [f(x) for x in x_values]

# Plot the function
plt.plot(x_values, y_values)
plt.title('Function with Jump Discontinuity')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
